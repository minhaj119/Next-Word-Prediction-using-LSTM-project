```python
# Create a simple text dataset for Next Word Prediction

text_data = """
Artificial Intelligence and Machine Learning are transforming the world.
Data Science is the future of decision making.
Python is a popular programming language for AI and Data Analytics.
Deep Learning uses neural networks to process complex data.
Automation and AI are changing industries globally.
Machine Learning helps systems learn from experience without being explicitly programmed.
Predictive analytics enables organizations to make data-driven decisions.
Artificial Intelligence is revolutionizing the way we work and interact with technology.
"""

# Save the text as a CSV file
with open("text_dataset.csv", "w") as file:
    file.write("text\n")
    for line in text_data.strip().split("\n"):
        file.write(f"{line.strip()}\n")

print("✅ text_dataset.csv file created successfully!")
```
